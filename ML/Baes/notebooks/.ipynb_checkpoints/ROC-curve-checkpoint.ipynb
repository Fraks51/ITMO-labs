```python
from typing import List, Any
import main
import os
from matplotlib import pyplot as plt

def draw_roc(x, y):
    plt.plot(x, y)
    plt.ylabel('TPR')
    plt.xlabel('FPR')
    plt.show()
```

```python
grams_sets = []
for i in range(1, 11):
    path = "../data/part" + str(i)
    part_set_of_grams = [main.made_n_gram(path, name, 2)
                            for name in os.listdir(path=path)]
    grams_sets = grams_sets + part_set_of_grams
model = main.BayesClassifier(0.0001, [1, 1])
model.fit(grams_sets)
predicted = model.binary_predict(grams_sets)
data_predict = []
m_zero, m_one = 0, 0
for i in range(len(grams_sets)):
    _, y = grams_sets[i]
    if y == 1:
        m_one += 1
    else:
        m_zero += 1
    data_predict.append((predicted[i], y))
data_predict.sort()
data_predict.reverse()
print(data_predict)
x, y = [0], [0]
for pred, clazz in data_predict:
    if clazz == 0:
        x.append(x[-1] + 1/m_zero)
        y.append(y[-1])
    else:
        x.append(x[-1])
        y.append(y[-1] + 1/m_one)
draw_roc(x, y)
```
