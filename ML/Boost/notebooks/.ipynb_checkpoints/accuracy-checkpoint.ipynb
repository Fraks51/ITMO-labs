```python
import main
from matplotlib import pyplot as plt

def draw_plot(x, y):
    plt.plot(x, y)
    plt.ylabel('Accuracy')
    plt.xlabel('Steps')
    plt.show()
```

```python
y, X = main.get_normalized_data("../data/chips.csv")
x_0 = []
y_0 = []
for i in range(1, 56, 5):
    x_0.append(i)
    y_0.append(main.get_accuracy(y, X, i))
draw_plot(x_0, y_0)
```
