```python
import main
import numpy as np
import matplotlib.pyplot as plt


def predict(xx, yy, f):
    xx_ = xx.ravel()
    yy_ = yy.ravel()
    z = []
    for i in range(len(xx_)):
        z.append(main.sign(f(xx_[i], yy_[i])))
    z = np.array(z)
    return z.reshape(xx.shape)


def draw_plot(x, y, clazz, alpha, K, train_x, train_y, b):
    x_ = np.arange(0, 24, 0.5)
    y_ = np.arange(0, 10, 0.1)
    xx, yy = np.meshgrid(x_, y_)
    f = lambda x1, x2: sum([alpha[i] * train_y[i] * K([x1, x2], train_x[i]) for i in range(len(train_x))]) + b
    z = predict(xx, yy, f)
    plt.contourf(xx, yy, z)
    plt.xlim(0, 24)
    plt.ylim(0, 10)
    plt.scatter(x, y, c=clazz, cmap='inferno')
    plt.show()
```

```python
import pandas as pd

dataset = pd.read_csv("../data/geyser.csv")
data_size = len(dataset)
data_values = dataset.replace('P', 1).replace('N', -1).values.tolist()
x, y = [], []
x_coord = []
y_coord = []
for i in data_values:
    y.append(i[2])
    x.append(i[0:2])
    x_coord.append(i[0])
    y_coord.append(i[1])
K = main.get_kernel("linear_kernel")
alphas, b = main.get_weights_final(main.get_kernel("linear_kernel"), len(x), 0.5, 5, x, y)

draw_plot(x_coord, y_coord, y, alphas, K, x, y, b)
```
